# helper functions

```python
def types_of_data(series):
    
    numbers, strings, blanks, n_null = 0, 0, 0, 0
    
    for item in series:
        try:
            float(item)
            if pd.isnull(item):
                n_null += 1
            else:
                numbers += 1
        except:
            strings += 1
            if str(item) == ' ' or str(item) == '':
                blanks += 1
                
    # note: blanks are strings
    return numbers, strings, blanks, n_null


def print_data_info(series):
    
    numbers, strings, blanks, n_null = types_of_data(series)
    
    print('UNIQUE numbers, strings, blanks, null =', types_of_data(series.unique()))
    print('number of numeric =', numbers)
    print('number of strings =', strings)
    print('number of blanks =', blanks)  # blanks are strings
    print('number of null = ', n_null)
    print('check: ', numbers + strings + n_null, '==', series.shape[0])


# test
# types_of_data(df_better.GONAD_WEIGHT.unique())
# types_of_data(df_better.GONAD_WEIGHT)
print_data_info(df_better.GONAD_WEIGHT)
```

```python
def types_of_data(series):
    numbers, strings, blanks = 0, 0, 0
    for item in list(series.unique()):
        try:
            float(item)
            numbers += 1
        except:
            if str(item) == ' ' or str(item) == '':
                blanks += 1
            elif pd.isnull(item):
                pass
            else:
                strings += 1
    return numbers, strings, blanks


def how_many_numeric(series):
    numbers = 0
    for item in series:
        try:
            float(item)
            if pd.isnull(item):
                pass
            else:
                numbers += 1
        except:
            pass
    return numbers


def how_many_strings(series):
    strings = 0
    for item in series:
        try:
            float(item)
        except:
            if pd.isnull(item):
                pass
            else:
                strings += 1
    return numbers


def how_many_blanks(series):
    return series[(series == ' ') | (series == '')].shape[0]


def print_data_info(series):
    n_numeric = how_many_numeric(series)
    n_string = how_many_strings(series)
    n_null = series.isnull().sum()
    
    print('UNIQUE numbers, strings, blanks =', types_of_data(series))
    print('number of numeric =', n_numeric)
    print('number of strings =', n_string)
    print('number of blanks =', how_many_blanks(series))  # blanks are strings
    print('number of null = ', n_null)
    print('check: ', n_numeric + n_string + n_null, '==', series.shape[0])
```
