```python
# there is only one item that we can call "blank", no strings, no variations

numbers = 0
strings = 0
blanks = 0

for item in list(df_raw.GONAD_WEIGHT.unique()):
    try:
        float(item)
        numbers += 1
    except:
        if str(item) == ' ' or str(item) == '':
            blanks += 1
        else:
            strings += 1
            
numbers, strings, blanks
```

```python
# all "blanks" are ' ' (there are no '')

numbers = 0
blanks = 0
blanks_v2 = 0

for item in list(df_raw.GONAD_WEIGHT.unique()):
    try:
        float(item)
        numbers += 1
    except:
        if str(item) == ' ':
            blanks += 1
        else:
            blanks_v2 += 1
            
numbers, blanks, blanks_v2
```
